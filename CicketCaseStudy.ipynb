```python
#playes playing the match
matches_played = pd.read_excel('IndiaStats.xlsx', sheetname='Sheet1') 
matches_played.head(5)
```

```python
#indian players
indian_match = matches_played[matches_played['Team'] == 'India']
indian_match.head(15)
```

```python
#Total Runs of Team India Team
total_Runs_India = sum(indian_match['Runs'])
total_Runs_India
```

```python
#Total Innings Not Out India Team
total_not_out = sum(indian_match["Not Outs"])
total_not_out
```

```python
#Total Innings Played India Team
total_innings = sum(indian_match["Bat_Innings"])
total_innings
```

```python
#Total Balls Faced India Team
total_ball_faced = sum(indian_match["Balls Faced"])
total_ball_faced
```

```python
#Total No 100s and 50s India Team
total_100_50 = sum(indian_match["50s"]) + sum(indian_match["100s"])
total_100_50
```

```python
# total boundary runs per innings of India
total_boundary = (4*sum(indian_match["4s"])+6*sum(indian_match["6s"]))/total_innings
total_boundary
```

```python
#Total Batting Average of Indian Team
batting_avg_india = total_Runs_India/(total_innings-total_not_out)
round(batting_avg_india,1)
```

```python
#Total Strike Rate India Team
strike_rate_india = (total_Runs_India/total_ball_faced)*100
strike_rate_india
```

```python
#mile stone reaaching capability India Team
mra_india =(total_100_50/total_innings)*100
mra_india
```

```python
#total outrate of India Team
outrate_india = (total_innings-total_not_out)*100/total_ball_faced
outrate_india
```

```python
#Total Runs of Australia Team
total_Runs_aus = sum(aus_match['Runs'])
total_Runs_aus
```

```python
#Total Innings Not Out Australia Team
total_not_out_aus = sum(aus_match["Not Outs"])
total_not_out_aus
```

```python
#Total Batting Average of Australia Team
batting_avg_aus = total_Runs_aus/(total_innings_aus-total_not_out_aus)
round(batting_avg_aus,1)
```

```python
#Total Strike Rate Australia Team
strike_rate_aus = (total_Runs_aus/total_ball_faced_aus)*100
strike_rate_aus
```

```python
#previous matches played
previous_matches = pd.read_excel('IndiaVsAus.xlsx', sheetname='Sheet1') 
previous_matches.head(5)
```

```python
#India Bowlers
bowler = matches_played[(matches_played['Primary'] == "bowler") | (matches_played['Primary'] == "batsman-bowler")]
india_bowler = bowler[matches_played['Team'] == "India"]
aus_bowler = bowler[matches_played['Team'] == "Australia"]
print(india_bowler)
print(aus_bowler)
```

```python
#balling average
runs_conceled_ind = sum(india_bowler['Runs Given'])
runs_conceled_aus = sum(aus_bowler['Runs Given'])
print(runs_conceled_ind)
print(runs_conceled_aus)
```

```python
#bowls bowled
bowls_ind = sum(india_bowler['Balls Bowled'])
bowls_aus = sum(aus_bowler['Balls Bowled'])
print(bowls_ind)
print(bowls_aus)
```

```python
# wickets taken
wicket_ind = sum(india_bowler['wickets'])
wicket_aus = sum(aus_bowler['wickets'])
print(wicket_ind)
print(wicket_aus)
```

```python
# overs bowled
over_bowl_ind = sum(india_bowler['overs bowled'])
over_bowl_aus = sum(aus_bowler['overs bowled'])
print(over_bowl_ind)
print(over_bowl_aus)
```

```python
#Bowl avg per wicket
ba_india = runs_conceled_ind/wicket_ind
ba_aus = runs_conceled_aus/wicket_aus
print(ba_india)
print(ba_aus)
```

```python
#Bowl strike rate
bsr_india = over_bowl_ind/wicket_ind
bsr_aus = over_bowl_aus/wicket_aus
print(bsr_india)
print(bsr_aus)
```

```python
#bowler economy
be_india = runs_conceled_ind/over_bowl_ind
be_aus = runs_conceled_aus/over_bowl_aus
print(be_india)
print(be_aus)
```

```python
#outrate bowler
ob_ind = wicket_ind/bowls_ind
ob_aus = wicket_aus/bowls_aus
print(ob_ind)
print(ob_aus)
```
